```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
pd.set_option('chained_assignment',None)
```

```python
log_data  = open('history_server.log', 'r')
split_list = []

for line in log_data:
    thing1 = line.split(' - ')
    date = thing1[0][:10]
    if len(thing1[0]) != 23:
        next(log_data)
        continue
    time = thing1[0][11:19]
    type_1 = thing1[1]
    type_2 = thing1[2]
    info = thing1[3]
    
    split_list.append([date, time, type_1, type_2, info])

df = pd.DataFrame(split_list, columns=['date', 'time', 'source', 'type', 'info'])
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
```

Passo tutta la riga splittando quando c'è "-". Successivamente costruisco il dataset di quello che contieene le domande qualora volessimo disegnare le occorrenze delle domande più fatte

```python
df_domande = df.loc[df['info'].str.contains("Replyed")]
df_domande[['info','domanda']] = df['info'].str.split(":", expand = True)
df_domande['domanda'] = df_domande['domanda'].str.replace(r"\n", "")
```

```python
df_domande
```

Qua invece creo quello dei conteggi per ogni tipo di cosa che rileva il bot

```python
df_count = df.groupby(["date", "type"]).count().reset_index()[['date', 'type', 'info']].rename(columns = {'info': 'count'})
df_count
```

Disegno infine il numero di domande, warning ed errori che dà il bot

```python
import matplotlib.dates as mdates

dati_info =  df_count.loc[df_count['type'] == "INFO"]

fig, ax = plt.subplots()
plt.plot(dati_info['date'], dati_info['count'], color = "#00a9c3", alpha = 0.8, linewidth =2)
l1 = plt.scatter(x = dati_info['date'], y = dati_info['count'], color = "#00a9c3", alpha = 1)

ax.xaxis.set_major_locator(mdates.DayLocator(interval = 1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%h'))
plt.ylim(bottom = 0)
plt.xlabel("Data", size = 12)
plt.ylabel("Conteggi", size = 12)
plt.xticks(size = 10)
plt.yticks(size = 10)
plt.title("DOMANDE")

lg = plt.legend([l1], ["{}: {}".format(dati_info['date'].iloc[-1].strftime("%d-%h"), int(dati_info['count'].iloc[-1]))], 
                                       bbox_to_anchor=(1.01, 0.6, 1.1, 0.2), loc='upper left')
plt.grid(alpha = 0.5)
#fig.savefig("pics/report/domande.png", dpi = 100, bbox_extra_artists=(lg,), bbox_inches='tight')
```

```python
dati_warning =  df_count.loc[df_count['type'] == "WARNING"]

fig, ax = plt.subplots()
plt.plot(dati_warning['date'], dati_warning['count'], color = "red", alpha = 0.8, linewidth =2)
l1 = plt.scatter(x = dati_warning['date'], y = dati_warning['count'], color = "red", alpha = 1)

plt.annotate(dati_warning['count'].iloc[-1],
             xy=(dati_warning['date'].iloc[-1],dati_warning['count'].iloc[-1]),
             xytext=(dati_warning['date'].iloc[-1],dati_warning['count'].iloc[-1]-5), ha='left')

ax.xaxis.set_major_locator(mdates.DayLocator(interval = 1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%h'))
plt.ylim(bottom = 0)
plt.xlabel("Data", size = 12)
plt.ylabel("Conteggi", size = 12)
plt.xticks(size = 10)
plt.yticks(size = 10)
plt.title("WARNING")

lg = plt.legend([l1], ["{}: {}".format(dati_warning['date'].iloc[-1].strftime("%d-%h"), int(dati_warning['count'].iloc[-1]))], 
                                       bbox_to_anchor=(1.01, 0.6, 1.1, 0.2), loc='upper left')
plt.grid(alpha = 0.5)
#fig.savefig("pics/report/warning.png", dpi = 100, bbox_extra_artists=(lg,), bbox_inches='tight')
```

```python
dati_error =  df_count.loc[df_count['type'] == "ERROR"]

fig, ax = plt.subplots()
plt.plot(dati_error['date'], dati_error['count'], color = "red", alpha = 0.8, linewidth =2)
l1 = plt.scatter(x = dati_error['date'], y = dati_error['count'], color = "red", alpha = 1)


ax.xaxis.set_major_locator(mdates.DayLocator(interval = 1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%h'))
plt.ylim(bottom = 0)
plt.xlabel("Data", size = 12)
plt.ylabel("Conteggi", size = 12)
plt.xticks(size = 10)
plt.yticks(size = 10)
plt.title("ERROR")

lg = plt.legend([l1], ["{}: {}".format(dati_error['date'].iloc[-1].strftime("%d-%h"), int(dati_error['count'].iloc[-1]))], 
                                       bbox_to_anchor=(1.01, 0.6, 1.1, 0.2), loc='upper left')
plt.grid(alpha = 0.5)
#fig.savefig("pics/report/error.png", dpi = 100, bbox_extra_artists=(lg,), bbox_inches='tight')
```
